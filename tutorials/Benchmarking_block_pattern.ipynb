```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
results_dir = 'Benchmark_data/block_pattern_data/'
```

```python
# ideal
SONAR_path_ideal = results_dir + 'SONAR/result_spot_all.csv'
cell2location_path_ideal = results_dir + 'cell2location/result_spot_all.csv'
Redeconve_path_ideal = results_dir + 'Redeconve/result_spot_all.csv'
DestVI_path_ideal = results_dir + 'DestVI/result_spot_all.csv'
DSTG_path_ideal = results_dir + 'DSTG/result_spot_all.csv'
GraphST_path_ideal = results_dir + 'GraphST/result_spot_all.csv'
df_SONAR_ideal = pd.read_csv(SONAR_path_ideal)
df_cell2location_ideal = pd.read_csv(cell2location_path_ideal)
df_Redeconve_ideal = pd.read_csv(Redeconve_path_ideal)
df_DestVI_ideal = pd.read_csv(DestVI_path_ideal)
df_DSTG_ideal = pd.read_csv(DSTG_path_ideal)
df_GraphST_ideal = pd.read_csv(GraphST_path_ideal)
```

```python
# ideal-missing
SONAR_path_missing = results_dir + 'SONAR/result_spot_all_-.csv'
cell2location_path_missing = results_dir + 'cell2location/result_spot_all_-.csv'
Redeconve_path_missing = results_dir + 'Redeconve/result_spot_all_-.csv'
DestVI_path_missing = results_dir + 'DestVI/result_spot_all_-.csv'
DSTG_path_missing = results_dir + 'DSTG/result_spot_all_-.csv'
GraphST_path_missing = results_dir + 'GraphST/result_spot_all_-.csv'
df_SONAR_missing = pd.read_csv(SONAR_path_missing)
df_cell2location_missing = pd.read_csv(cell2location_path_missing)
df_Redeconve_missing = pd.read_csv(Redeconve_path_missing)
df_DestVI_missing = pd.read_csv(DestVI_path_missing)
df_DSTG_missing = pd.read_csv(DSTG_path_missing)
df_GraphST_missing = pd.read_csv(GraphST_path_missing)
```

```python
# real
SONAR_path_real = results_dir + 'SONAR/result_spot_all_real.csv'
cell2location_path_real = results_dir + 'cell2location/result_spot_all_real.csv'
Redeconve_path_real = results_dir + 'Redeconve/result_spot_all_real.csv'
DestVI_path_real = results_dir + 'DestVI/result_spot_all_real.csv'
DSTG_path_real = results_dir + 'DSTG/result_spot_all_real.csv'
GraphST_path_real = results_dir + 'GraphST/result_spot_all_real.csv'
df_SONAR_real = pd.read_csv(SONAR_path_real)
df_cell2location_real = pd.read_csv(cell2location_path_real)
df_Redeconve_real = pd.read_csv(Redeconve_path_real)
df_DestVI_real = pd.read_csv(DestVI_path_real)
df_DSTG_real = pd.read_csv(DSTG_path_real)
df_GraphST_real = pd.read_csv(GraphST_path_real)
```

```python
# free
BayesTME_path = results_dir + 'BayesTME/result_spot_all.csv'
STdeconvolve_path = results_dir + 'STdeconvolve/result_spot_all.csv'
CFS_path = results_dir + 'CFS/result_spot_all.csv'
SURF_path =results_dir + 'SURF/result_spot_all.csv'
df_BayesTME = pd.read_csv(BayesTME_path)
df_STdeconvolve = pd.read_csv(STdeconvolve_path)
df_CFS = pd.read_csv(CFS_path)
df_SURF = pd.read_csv(SURF_path)
```

```python
df_value = pd.concat((df_SONAR_ideal['PCC'], df_cell2location_ideal['PCC'], df_Redeconve_ideal['PCC'], df_DSTG_ideal['PCC'], df_DestVI_ideal['PCC'], df_GraphST_ideal['PCC'],
                      df_SONAR_missing['PCC'], df_cell2location_missing['PCC'], df_Redeconve_missing['PCC'], df_DSTG_missing['PCC'], df_DestVI_missing['PCC'], df_GraphST_missing['PCC'],
                      df_SONAR_real['PCC'], df_cell2location_real['PCC'], df_Redeconve_real['PCC'], df_DSTG_real['PCC'], df_DestVI_real['PCC'], df_GraphST_real['PCC'],
                      df_CFS['PCC'], df_STdeconvolve['PCC'], df_BayesTME['PCC'], df_SURF['PCC']), axis=0)
df_value.columns = ['PCC']
df_value.reset_index(drop=True, inplace=True)
df_method = pd.DataFrame(['SONAR0'] * len(df_SONAR_ideal) + ['Cell2location0'] * len(df_cell2location_ideal) + ['Redeconve0'] * len(df_Redeconve_ideal) + ['DSTG0'] * len(df_DSTG_ideal) + ['DestVI0'] * len(df_DestVI_ideal) + ['GraphST0'] * len(df_GraphST_ideal) +
                         ['SONAR1'] * len(df_SONAR_missing) + ['Cell2location1'] * len(df_cell2location_missing) + ['Redeconve1'] * len(df_Redeconve_missing) + ['DSTG1'] * len(df_DSTG_missing) + ['DestVI1'] * len(df_DestVI_missing) +['GraphST1'] * len(df_GraphST_missing) +
                         ['SONAR2'] * len(df_SONAR_real) + ['Cell2location2'] * len(df_cell2location_real) + ['Redeconve2'] * len(df_Redeconve_real) + ['DSTG2'] * len(df_DSTG_real) + ['DestVI2'] * len(df_DestVI_real) + ['GraphST2'] * len(df_GraphST_real) +
                         ['CFS'] * len(df_CFS) + ['STdeconvolve'] * len(df_STdeconvolve) + ['BayesTME'] * len(df_BayesTME) + ['SURF'] * len(df_SURF), columns=['method'])
df_PCC = pd.concat([df_value, df_method], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(4, 6))
sns.boxplot(y='method', x='PCC', data=df_PCC, width=0.5, showfliers=False, whis=0.5, orient='h')
ax.set_yticklabels(['SONAR', 'Cell2location', 'Redeconve', 'DSTG', 'DestVI', 'GraphST',
                    'SONAR', 'Cell2location', 'Redeconve', 'DSTG', 'DestVI', 'GraphST',
                    'SONAR', 'Cell2location', 'Redeconve', 'DSTG', 'DestVI', 'GraphST',
                    'CFS', 'STdeconvolve', 'BayesTME', 'SURF'])
plt.subplots_adjust(left=0.3, right=0.9, top=0.9, bottom=0.1)
plt.xticks(fontsize=12, fontname='Arial')
plt.yticks(fontsize=12, fontname='Arial')
plt.xlabel('PCC', family='Arial', fontsize=12)
plt.ylabel('', family='Arial', fontsize=12)
plt.gca().xaxis.labelpad = 10
plt.show()
```

```python
df_value = pd.concat((df_SONAR_ideal['RMSE'], df_cell2location_ideal['RMSE'], df_Redeconve_ideal['RMSE'], df_DSTG_ideal['RMSE'], df_DestVI_ideal['RMSE'], df_GraphST_ideal['RMSE'],
                      df_SONAR_missing['RMSE'], df_cell2location_missing['RMSE'], df_Redeconve_missing['RMSE'], df_DSTG_missing['RMSE'], df_DestVI_missing['RMSE'], df_GraphST_missing['RMSE'],
                      df_SONAR_real['RMSE'], df_cell2location_real['RMSE'], df_Redeconve_real['RMSE'], df_DSTG_real['RMSE'], df_DestVI_real['RMSE'], df_GraphST_real['RMSE'],
                      df_CFS['RMSE'], df_STdeconvolve['RMSE'], df_BayesTME['RMSE'], df_SURF['RMSE']), axis=0)
df_value.columns = ['RMSE']
df_value.reset_index(drop=True, inplace=True)
df_method = pd.DataFrame(['SONAR0'] * len(df_SONAR_ideal) + ['Cell2location0'] * len(df_cell2location_ideal) + ['Redeconve0'] * len(df_Redeconve_ideal) + ['DSTG0'] * len(df_DSTG_ideal) + ['DestVI0'] * len(df_DestVI_ideal) + ['GraphST0'] * len(df_GraphST_ideal) +
                         ['SONAR1'] * len(df_SONAR_missing) + ['Cell2location1'] * len(df_cell2location_missing) + ['Redeconve1'] * len(df_Redeconve_missing) + ['DSTG1'] * len(df_DSTG_missing) + ['DestVI1'] * len(df_DestVI_missing) + ['GraphST1'] * len(df_GraphST_missing) +
                         ['SONAR2'] * len(df_SONAR_real) + ['Cell2location2'] * len(df_cell2location_real) + ['Redeconve2'] * len(df_Redeconve_real) + ['DSTG2'] * len(df_DSTG_real) + ['DestVI2'] * len(df_DestVI_real) + ['GraphST2'] * len(df_GraphST_real) +
                         ['CFS'] * len(df_CFS) + ['STdeconvolve'] * len(df_STdeconvolve) + ['BayesTME'] * len(df_BayesTME) + ['SURF'] * len(df_SURF), columns=['method'])
df_RMSE = pd.concat([df_value, df_method], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(4, 6))
sns.boxplot(y='method', x='RMSE', data=df_RMSE, width=0.5, showfliers=False, whis=0.5, orient='h')
ax.set_yticklabels(['SONAR', 'Cell2location', 'Redeconve', 'DSTG', 'DestVI', 'GraphST',
                    'SONAR', 'Cell2location', 'Redeconve', 'DSTG', 'DestVI', 'GraphST',
                    'SONAR', 'Cell2location', 'Redeconve', 'DSTG', 'DestVI', 'GraphST',
                    'CFS', 'STdeconvolve', 'BayesTME', 'SURF'])
plt.subplots_adjust(left=0.3, right=0.9, top=0.9, bottom=0.1)
plt.xticks(ticks=np.arange(0, 0.4, 0.1), fontsize=12, fontname='Arial')
plt.yticks(fontsize=12, fontname='Arial')
plt.xlabel('RMSE', family='Arial', fontsize=12)
plt.ylabel('', family='Arial', fontsize=12)
plt.gca().xaxis.labelpad = 10
plt.show()
```
